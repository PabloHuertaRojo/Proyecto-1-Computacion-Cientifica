```python
#Problema 1: Parte 1

import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return x- np.abs(x)**(4/5)+2

def f2(x):
    return x- np.exp(-x)+4

def g1(x):
    return np.abs(x)**(4/5)-2

def g2(x):
    return -1*np.log(x+4)


n = 100
x0_1 = -2
xn_1 = 2
dif_1 = xn_1 - x0_1
x0_2 = -2
xn_2 = 2
dif_2 = xn_2 - x0_2

abscisa=np.arange(1,n+1)


x_1 = np.zeros(n)
x_2 = np.zeros(n)
y_1 = np.zeros(n)
y_2 = np.zeros(n)
f_1 = np.zeros(n)
f_2 = np.zeros(n)

k_1=0
k_2=0
for i in range(n):
    k_1=x0_1 + dif_1*i/(n-1)
    k_2=x0_2 + dif_2*i/(n-1)
    x_1[i]=k_1
    x_2[i]=k_2
    y_1[i]=g1(k_1)
    y_2[i]=g2(k_2)
    f_1[i]=f1(k_1)
    f_2[i]=f2(k_2)
    

plt.figure(1)
plt.plot(x_1,y_1)
plt.plot(x_1,x_1)

plt.figure(3)
plt.plot(x_2,y_2)
plt.plot(x_2,x_2)

plt.figure(4)
plt.plot(x_1,f_1)

plt.figure(5)
plt.plot(x_2,f_2)
```
