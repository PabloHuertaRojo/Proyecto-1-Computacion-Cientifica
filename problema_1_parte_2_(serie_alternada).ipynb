```python
#Problema 1: Parte 2.1

import numpy as np
import matplotlib.pyplot as plt

s_sol=-np.log(2)
s1=0
s=[]
s.append(s1)

x=[]
x.append(0)

tol=0.0001
count=1
while(np.abs(s1-s_sol)>tol):
    
    s2=s1+(-1)**count/count
    
    s1=s2
    
    s.append(s1)
    x.append(count)
    
    count=count+1

    
print('la solución es')
print(s_sol)
print('la aproximcación es')
print(s1)
print('número de iteraciones')
print(count)
```

```python
#Problema 1: Parte 2.2

import numpy as np
import matplotlib.pyplot as plt

def Altern(n):
    M=np.zeros((n,n))
    N=np.zeros((n,n))
    for i in range(0,n):
        for j in range(0,n):
            if(j==0):
                N[i,j]=1
            if (j>0) and (i>0):
                N[i,j]=N[i-1,j]+N[i-1,j-1]
            M[i,j]=N[i,j]/(2**i)
    return M
    

n=4
k=6

sol=-np.log(2)
nt=n+k
S=np.zeros(nt)

s1=0
for i in range(1,nt+1):
    s1+=(-1)**i/i
    S[i-1]=s1
    
A=Altern(k)

M=0
for i in range(k):
    M+=S[n+i]*A[n+1,i]
    print(i)
    print(M)

error=np.abs(M-sol)

print('el error es')
print(error)

print('A')
print(A)
```

```python
#Problema 1: Parte 2.4

import numpy as np
import matplotlib.pyplot as plt
    
n=5
k=4

nk=1+2*k
nt=n+nk
sol=np.pi
S=np.zeros(nk)
A=np.zeros((nk,k+1))
Er=np.zeros((nk,k+1))

s1=0
for i in range(nt):
    s1+=4*(-1)**i/(2*i+1)
    if i>n-1:
        S[i-n]=s1

for i in range(k+1):
    for j in range(nk-2*i):
        if i==0:
            A[j,i]=S[j]
            Er[j,i]=np.abs(S[j])-np.abs(sol)
        if i>0:
            A[j,i]=A[j+2,i-1]-(A[j+2,i-1]-A[j+1,i-1])**2/((A[j+2,i-1]-A[j+1,i-1])-(A[j+1,i-1]-A[j,i-1]))
            Er[j,i]=np.abs(A[j,i])-np.abs(sol)

print(A) #La primera columna son los valores de S, la segunda columna es la primera transformación de Aitken, la tercera es la segunda transformación, y así
print(Er) #Son los errores de la matriz A con respecto a la solución
print(sol)
```
