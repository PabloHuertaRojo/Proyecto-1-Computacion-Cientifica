```python
# Problema 1: Parte 3.2

import numpy as np
import matplotlib.pyplot as plt

def f1(n,n1):
    k=0
    nt=n+n1
    for i in range(nt):
        k+=1/(i+1)-np.log(1+1/(i+1))
    return k

def Richardson(n,n1,f):
    v=np.zeros((n,n))
    v0=np.zeros(n)
    
    for j in range(n):
        l=2**(j+1)
        v0[j]=f(l,n1)
    
    for i in range(n):
        for j in range(n):
            
            if i==0:
                
                v[i,j]=v0[j]
                
            if (i==1) and (j<n-i):
                
                v[i,j]=(2**(i)*v0[j+1]-v0[j])/(2**i-1)
                
            if (i>1) and (j<n-i):
                
                v[i,j]=(2**(i)*v[i-1,j+1]-v[i-1,j])/(2**i-1)
                
    return(v)
    #return v[n-1,0]

#---------------------------------------------------------------------------------------------------

n=4
n1=5
tol=10**(-6)

Mv=Richardson(n,n1,f1)
print(Mv)

sol=0.577215664901532

count=0
k=0
while(np.abs(sol-k)>tol):
    count+=1
    k+=1/(count)-np.log(1+1/(count))

n0=count
print('el error sin usar Richardson es')
print(np.abs(k-sol))
print('el número de iteraciones usadas es de')
print(n0)
    
R=0
count=1
tol=10**(-6)
while(np.abs(sol-R)>tol):
    count+=1
    MR=Richardson(count,n1,f1)
    R=MR[count-1,0]
    
tR=count
print('El error usando Richardson es de')
print(np.abs(R-sol))
print('número de transformaciones utilizadas')
print(tR)
```

```python
# Problema 1: Parte 3.3

import numpy as np
import matplotlib.pyplot as plt

def w_func(n,n1):
    c=0    #c1
    w=2    #w1
    nt=n+n1
    for i in range(nt):
        c=np.sqrt((1+c)/2)
        w=w/c
    return w

def Richardson(n,n1,f):
    v=np.zeros((n,n))
    v0=np.zeros(n)
    
    for j in range(n):
        l=2**(j+1)
        v0[j]=f(l,n1)
    
    for i in range(n):
        for j in range(n):
            
            if i==0:
                
                v[i,j]=v0[j]
                
            if (i==1) and (j<n-i):
                
                v[i,j]=(2**(i)*v0[j+1]-v0[j])/(2**i-1)
                
            if (i>1) and (j<n-i):
                
                v[i,j]=(2**(i)*v[i-1,j+1]-v[i-1,j])/(2**i-1)
                
    return(v)

#--------------------------------------------------------------------------------------

n=3
n1=20
tol=10**(-14)

sol=3.141592653589793

R=0
count=1
while(np.abs(sol-R)>tol):
    count+=1
    MR=Richardson(count,n1,w_func)
    R=MR[count-1,0]
    print(sol-R)
    
tR=count
print('El error usando Richardson es de')
print(np.abs(R-sol))
print('número de transformaciones utilizadas')
print(tR)
```
